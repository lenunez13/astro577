```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



```

```python
file = 'all_planets.csv'
df = pd.read_csv(file,comment="#")
td = "pl_trandep"
disc = "pl_disc"


plt.rcParams.update({'font.size': 18})
fig, ax1 = plt.subplots()
ax1.plot(df[disc],df[td],'o',color='k')
ax1.set_ylabel('Transit Depth (percent)')
ax1.set_xlabel('Year of Discovery')


```
